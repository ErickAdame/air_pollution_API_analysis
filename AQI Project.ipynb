```python
#Dependencies
import matplotlib.pyplot as plt
import requests
import math
from scipy import stats
import pandas as pd
from config import weather_api_key
import numpy as np
from datetime import datetime
```

```python
#Input for city so this can be run as a program
city= input("Type a city to find AQI analysis: ").lower()
#Based on city, use API to gather lat and long
#URL is assembled
geo_url = "http://api.openweathermap.org/geo/1.0/direct?q=" + city + "&appid=" + weather_api_key
lat_lon = requests.get(geo_url).json()
print(geo_url)
```

```python
#Extract lat and lon
lat = str(lat_lon[0]["lat"])
lon = str(lat_lon[0]["lon"])
```

```python
#Creating URL for Air Pollution API
url = "http://api.openweathermap.org/data/2.5/air_pollution/history"
units = "imperial"

#Input for year OR season to analyze
year = input("Please enter a value to analyze: ").lower()

#Logic if running as a program
if year == "2022":
    time_in_out = "&start=1641013200&end=1672549140"
if year == "2021":
    time_in_out = "&start=1609477200&end=1641013140"
if year == "winter":
    time_in_out = "&start=1638334800&end=1646110740"
if year == "spring":
    time_in_out = "&start=1646110800&end=1654055940"
if year == "summer":
    time_in_out = "&start=1654056000&end=1662004740"
if year == "fall":
    time_in_out = "&start=1662004800&end=1669870740"
print("--------------------------")
print("All set, working on that now.....")

#Final URL to extract data
final_url = url + "?lat=" + lat + "&lon=" + lon + time_in_out + "&appid=" + weather_api_key
print(final_url)
```

```python
#Get request and switch to json format
aqi_num = requests.get(final_url).json()
```

```python
#Empty lists, later used to create dataframe
aqi = []
time_stamp = [] #raw time_stamp
ts = [] #reformatted time_stamp

#Loop to go through entire json to collect AQI and time_stamp
for day in range(len(aqi_num['list'])):
    aqi.append(aqi_num['list'][day]['main']['aqi'])
    time_stamp.append(aqi_num['list'][day]['dt'])
    
#Loop to reformat time stamp    
for time in time_stamp:
    ts.append(datetime.fromtimestamp(time).strftime("%m-%d-%Y"))
```

```python
#AQI Dataframe
aqi_df = pd.DataFrame({
    "Human Time": ts,
    "AQI": aqi
})

#Human Time was a string, switched to a datetime type to resample by Days
aqi_df["Human Time"] = pd.to_datetime(aqi_df["Human Time"], format="%m-%d-%Y")

#Resample/Group by days and rename data frame
aqi_groups = aqi_df.resample('D', on='Human Time').max()


#Gather value counts AQI per day and sort values in order 1 through 5
aqi_sorted = aqi_groups["AQI"].value_counts().sort_index()

#CSV for later use and other analyses
aqi_days_df = pd.DataFrame(aqi_sorted)


#If no AQI values exist of a certain number, must find a way to have ZERO placeholder.
#This workaround was used to solve this.
#A dataframe with value count of ZERO for each category is used to concat and create Nan values
#Nan values are then filled with ZERO. Placeholder is removed.
holder_df = pd.DataFrame({
    "AQI":[0,0,0,0,0,0]})
aqi_groups = pd.concat([holder_df,aqi_days_df], axis=1).fillna(0).iloc[:,1][1:].astype(int)


#Annual_df is resampled by week. Average AQI per week is used to look at seasonal patterns
annual_df = aqi_df.resample('w', on='Human Time').mean()



csv_df = pd.DataFrame(aqi_groups).reset_index()
lat_df = {"AQI": lat}
lon_df = {"AQI": lon}

csv_df = csv_df.append(lat_df, ignore_index = True)
csv_df = csv_df.append(lon_df, ignore_index = True)
csv_df['index'].loc[5] = "Lat"
csv_df['index'].loc[6] = "Lon"
csv_df.to_csv(f'./Resources/{city}_AQI_{year}.csv', index=False)
```

```python
#Seasonal Plot

city=city.title()
annual_df.plot(linewidth=4, legend=False, color="orange")
plt.ylim(0,5)
plt.title((f"{city} AQI Values ({year})"), size = 13)
plt.ylabel("Air Quality Index", size=13)
plt.xlabel("Month", size=13)
plt.savefig(f"{year}_{city}_AQI_LineGraph.png", transparent=True)
plt.show()
```

```python
#Bar graph

#AQI Values on X axis, value counts on the Y axis
x = [1,2,3,4,5]
y = aqi_groups

#Colors and Labels follow EPA usage for respective AQI values
labels = ['Good', 'Fair', 'Moderate', 'Poor', 'Very Poor']
colors = ['green','yellow','orange','red', 'purple']

barplot = plt.bar(x, y, color=colors, align="center", edgecolor="black")
plt.bar_label(barplot, labels=y, label_type="edge", padding=3)
x_axis = np.arange(len(x))
plt.ylim(0,210)
plt.ylabel("Number of Days", size=13)
plt.xlabel("Air Quality", size =13)
tick_locations = [value for value in x]
plt.xticks(tick_locations, ['Good', 'Fair', 'Moderate', 'Poor', 'Very Poor'])
plt.title((f"Overall Air Quality for {city} ({year})").title(), size = 13)
print("--------------------------")
print("A bar graph has been created with your AQI data")
plt.savefig(f"{city}_AQI_{year}.png", transparent=True)
plt.show()
```
